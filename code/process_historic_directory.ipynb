```python
import pandas as pd

from src.data_load import load_tables, load_instance, load_distances
from src.experimentation_config import instance_map

data_path = '../data'

instance = 'instRS1'

directorio_df, labors_raw_df, cities_df, duraciones_df, valid_cities = load_tables(data_path, generate_labors=False)
labors_real_df = load_instance(data_path, instance, labors_raw_df)
dist_dict = load_distances(data_path, 'osrm', instance)
```

```python
labors_real_df['date'] = labors_real_df['schedule_date'].dt.date
```

```python
columns = ['date','city', 'alfred', 'address_id', 'address_point']
labors_real_df = labors_real_df[labors_real_df['alfred'] != '']
hist_directory = labors_real_df.dropna(subset=['alfred']).drop_duplicates(columns).sort_values(['date', 'city', 'alfred']).reset_index(drop=True)[columns]
hist_directory['alfred'] = hist_directory['alfred'].astype(int).astype(str)
```

```python
hist_directory.to_csv(f'{data_path}/instances/{instance_map[instance]}_inst/{instance}/directorio_hist_df.csv',index=False)
```
