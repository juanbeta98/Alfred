```python
def merge_distance_dicts(dict_a, dict_b):
    merged = {}

    # Union of all main keys
    all_primary_keys = set(dict_a) | set(dict_b)

    for key in all_primary_keys:
        # get nested dicts (or empty if missing)
        sub_a = dict_a.get(key, {})
        sub_b = dict_b.get(key, {})

        # combine nested dicts
        merged_sub = {**sub_a, **sub_b}

        # filter out non-floats
        merged_sub = {
            k: v for k, v in merged_sub.items() if isinstance(v, float)
        }

        merged[key] = merged_sub

    return merged

a_dist_dict = merge_distance_dicts(a4_dist_dict, b4_dist_dict)

```
