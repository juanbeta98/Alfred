```python
import pandas as pd
import plotly.express as px

def plot_order_timing_histograms(df, created_col='created_at', schedule_col='schedule_date', unit='hours'):
    """
    Generates two Plotly histograms:
    1. Distribution of order creation times (time of day, 30-min bins)
    2. Distribution of time difference between creation and schedule dates
    
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame containing datetime columns.
    created_col : str
        Name of the column representing creation timestamps.
    schedule_col : str
        Name of the column representing scheduled timestamps.
    unit : str, default 'hours'
        Unit for time difference histogram ('hours' or 'minutes').
    """
    
    # --- Safety checks ---
    if created_col not in df.columns or schedule_col not in df.columns:
        raise ValueError(f"Columns '{created_col}' and/or '{schedule_col}' not found in dataframe.")
    if unit not in ['hours', 'minutes']:
        raise ValueError("unit must be either 'hours' or 'minutes'")

    # --- First plot: creation time of day ---
    created_local = df[created_col].dt.tz_convert(df[created_col].dt.tz)  # ensure tz-aware
    created_hours = created_local.dt.hour + created_local.dt.minute / 60  # fractional hour of day
    
    fig1 = px.histogram(
        x=created_hours,
        nbins=48,  # 24 hours * 2 bins per hour = 30 min bins
        labels={'x': 'Hour of Day (local time)', 'y': 'Number of Orders'},
        title='Distribution of Order Creation Times (Local Time)',
        color_discrete_sequence=['#0083B0'],
    )
    fig1.update_layout(
        bargap=0.05,
        xaxis=dict(
            tickmode='array',
            tickvals=list(range(0, 25, 2)),
            title='Hour of Day (024)',
        ),
        yaxis_title='Count',
        template='plotly_white',
    )

    # --- Second plot: time delta between schedule and creation ---
    delta = df[schedule_col] - df[created_col]
    delta_hours = delta.dt.total_seconds() / 3600
    delta_vals = delta_hours if unit == 'hours' else delta_hours * 60
    
    fig2 = px.histogram(
        x=delta_vals,
        nbins=50,
        labels={'x': f'Time Between Creation and Schedule ({unit})', 'y': 'Number of Orders'},
        title=f'Distribution of Time Between Creation and Schedule ({unit})',
        color_discrete_sequence=['#FF6B6B'],
    )
    fig2.update_layout(
        bargap=0.05,
        template='plotly_white',
    )

    return fig1, fig2
```

```python
fig1, fig2 = plot_order_timing_histograms(labors_real_df, created_col='created_at', schedule_col='schedule_date', unit='hours')

fig1.show()
fig2.show()
```
